```python
from bs4 import BeautifulSoup
import requests,openpyxl


excel=openpyxl.Workbook()
print(excel.sheetnames)
sheet=excel.active
sheet.title='Top Rated Movies'
print(excel.sheetnames)
sheet.append(['Movie Rank','Movie Name','Year Of Release','IMDB Rating'])



headers={'User-Agent':'Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.3 Mobile/15E148 Safari/604.1'}
try:
      source=requests.get('https://www.imdb.com/chart/top/',headers=headers)
      source.raise_for_status()

      soup=BeautifulSoup(source.text,'html.parser')
     # print(soup)
      movies=soup.find('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-between sc-3f13560f-0 sTTRj compact-list-view ipc-metadata-list--base").find_all('li')
     # print(len(movies))
      for movie in movies:

        name = movie.find('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b51a3d33-7 huNpFl cli-title").a.get_text(strip=True).split('.')[1]
        rank= movie.find('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b51a3d33-7 huNpFl cli-title").get_text(strip=True).split('.')[0]
        year= movie.find('div',class_="sc-b51a3d33-5 ibuRZu cli-title-metadata").span.text
        rating=movie.find('div',class_="sc-e3e7b191-0 iKUUVe sc-b51a3d33-2 ccUQup cli-ratings-container").span.get_text(strip=True).split('(')[0]
        print(rank,name,year,rating)
        sheet.append([rank,name,year,rating])


except Exception as e:
      print(e)


excel.save('IMDB Movie Rating.xlsx')
```
